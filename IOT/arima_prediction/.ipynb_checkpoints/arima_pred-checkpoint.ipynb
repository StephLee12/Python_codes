# ARIMA预测

## 数据可视化

```python
import pandas as pd
import matplotlib.pyplot as plt 

# load train dataset
df = pd.read_csv('./dataset/flood_data.csv',usecols=[1],engine='python',skipfooter=3)
df.plot()
```

## 数据的自相关性和偏相关性

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
# 画出自相关函数图
plot_acf(df['waterlevel'],lags=100)
plt.show()
```

```python
# 画出偏相关函数图
plot_pacf(df['waterlevel'])
plt.show()
```

```python
# 检测序列的平稳性
from statsmodels.tsa.stattools import adfuller
adfuller(df['waterlevel'])
```

```python
# 进行差分处理
diff_df = df.diff().dropna()
diff_df.plot()
```

```python
# 差分后的 自相关
plot_acf(diff_df)
plt.show()
```

```python
# 差分后的 偏相关
plot_pacf(diff_df)
plt.show()
```

```python
# 检测差分后平稳性
adfuller(diff_df['waterlevel'])
```
