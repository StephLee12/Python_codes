```python
import numpy as np
```

```python
# 电动骑车和道路行驶条件参数
theta = 15.0*np.pi/180 # 坡度 单位 rad
f = 0.015 # 地面滚动摩擦系数 
r = 0.25 # 车轮半径 单位 米
m = 1450.0 # 车的总质量 单位 千克
C_D = 0.28 # 迎风阻力系数
A = 1.95 # 车身横截面积 单位 平方米
k = 3.1 # 车轮到电机传动系统的变比
rho_air = 1.2258 # 空气密度
g = 9.8 # 重力加速度

# 电机参数
P_N = 40000.0 # 额定功率 单位 瓦
U_N = 400.0 # 额定电压 单位 伏
n_N = 5000.0 # 额定转速 单位 rpm
R_a = 0.1 # 电枢绕组电阻 单位 欧姆
I_aN = 110 # 额定电流 单位 安培
```

```python
# 计算汽车受到的阻力
#F_air = 0.5*C_D*A*rho_air*
F_f = m*g*np.cos(theta)*f # 地面的滚动阻力
F_G = m*g*np.sin(theta) # 坡道阻力
print('坡道阻力为{}N,地面的滚动阻力为{}N'.format(F_f,F_G))
```

```python
# 计算电动机的额定输出转矩
# 由于不计所有损耗 故电磁转矩即为输出转矩
E_aN = U_N-I_aN*R_a # 额定电枢电动势
P_emN = E_aN*I_aN # 额定电枢功率
T_outN = P_emN * 60 / (2*np.pi*n_N) # 额定电磁转矩
print('额定电枢电动势为{}V,额定输出转矩为{}N·m'.format(E_aN,T_outN))
```

```python
# 计算在电枢电压为E_a时的转速和输出转矩
def calc_n_T(E_a):
    n = E_a*n_N/E_aN # 转速
    I_a = (U_N-E_a)/R_a # 电枢电流
    #P_em = E_a*I_a # 电枢功率
    T_out = I_a*T_outN/I_aN # 电磁转矩
    return n,T_out
```

```python
# 计算车的速度
def calc_velocity(T):
    T_car = T*k
    F_total = T_car/(r) 
    F_air = F_total-F_f-F_G
    velocity = np.sqrt(F_air*2/(C_D*A*rho_air)) 
    return velocity*3.6/k
```

```python
for E_a in range(300,501):
    n,T_out = calc_n_T(E_a)
    velocity = calc_velocity(T_out)
    print('电枢电动势{}V,电磁转矩{}N·m,汽车速度{}km/h'.format(E_a,T_out,velocity))
```
